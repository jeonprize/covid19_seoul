```python
df = pd.read_csv("seoul_covid19.csv")
df
```

```python
df["거주지"].unique()
```

```python
df["타지역"] = df["거주지"].str.contains("기타") + df["거주지"].str.contains("타시도") + df["거주지"].str.contains("내발산동") 
df
```

```python
df["퇴원"] = df["조치상황"].str.contains("퇴원")
df["사망"] = df["조치상황"].str.contains("사망")
df["병원"] = df["조치상황"].str.contains("치료중")
df
```

```python
print(df["퇴원"].value_counts())
df["퇴원"].value_counts(normalize=True)
```

```python
print(df["사망"].value_counts())
df["사망"].value_counts(normalize=True)
```
