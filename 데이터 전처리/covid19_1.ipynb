```python
import re
def extract_num(num_string):
    if type(num_string) == str:
        num_string = num_string.replace("corona19","")
        sub_num = re.sub("[^0-9]","",num_string)
        sub_num = int(sub_num)
        return sub_num
    else:
        return num_string
```

```python
num_string = "<p class='corona19_no'>1</p>"
extract_num(num_string)
```

```python
df_all["연번"] = df_all["연번"].map(extract_num)
df_all["연번"].head()
```
