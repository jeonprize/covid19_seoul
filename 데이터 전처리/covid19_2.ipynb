```python
import pandas as pd
```

```python
df = pd.read_csv("seoul_covid19.csv")
df
```

```python
df = df.rename({"퇴원현황":"조치상황"},axis = "columns")
```

```python
df["확진일"] = pd.to_datetime(df["확진일"])
df["년월"] = df["확진일"].dt.strftime("%y-%m")
df["년월일"] = df["확진일"].dt.strftime("%y-%m-%d")
df["월"] = df["확진일"].dt.month
df["주"] = df["확진일"].dt.week
df["년주"] = df["확진일"].dt.strftime("%Y") +"-"+df["주"].astype(str)
df[["년월","년월일","월","주","년주"]]
```
