```python
import time
from tqdm import trange
```

```python
def get_multi_page_list(start_page, end_page):
    
    page_list = []
    for page_num in trange(start_page, end_page + 1):
        one_page = get_seoul_covid19_100(page_num)
        if len(one_page["data"]) > 0 :
            one_page = pd.DataFrame(one_page["data"])
            page_list.append(one_page)
            time.sleep(0.7)
        else :
            return page_list
    return page_list
```
